## Divergence and Overflow
### $h(x) = x^5 - 2x^4 + 2$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# make data
x_3 = np.linspace(start=-2.5, stop=2.5, num=100)

def h(x):
    return x**5 - 2*x**4 + 2

def dh(x):
    return 5*x**4 - 8*x**3
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    
    for n in range(500):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient
        
        step_size = abs(new_x - previous_x)
        x_list.append(new_x) 
        slope_list.append(derivative_func(new_x))
        
        if step_size < precision:
            break
            
    return new_x, x_list, slope_list
```

```python
# calling gradient descent function
local_min, list_x, deriv_list = gradient_descent(derivative_func=dh, initial_guess=0.2)

# plot function, derivative and scatter plot side by side 
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)

plt.xlim([-1.2, 2.5])
plt.ylim(-1, 4)
plt.title("Cost function")
plt.xlabel("X", fontsize=14)
plt.ylabel("h(x)", fontsize=14)

plt.plot(x_3, h(x_3))
plt.scatter(list_x, h(np.array(list_x)), color='red', s=100, alpha=0.6)

plt.subplot(1, 2, 2)
plt.xlim([-1, 2])
plt.ylim(-4, 5)
plt.title("Slope of the cost function")
plt.xlabel("X", fontsize=14)
plt.ylabel("dh(x)", fontsize=14)
plt.grid()

plt.plot(x_3, dh(x_3))
plt.scatter(list_x, deriv_list, color='red', s=100, alpha=0.6)

plt.show()

print('Local min orrucs at', local_min)
print('Cost at this minimum is:', h(local_min))
print('Number of steps:', len(list_x))
```
